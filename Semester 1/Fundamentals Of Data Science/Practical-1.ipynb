```python
# 1. DataFrame
'''Pandas DataFrame is a widely used data structure which works with a 
two-dimensional array with labeled axes (rows and columns). '''

import pandas as pd

mydic = {
    'name': ['a', 'b', 'c', 'd', 'e', 'f', 'g'],
    'age': [22, 23, 24, 23, 25, 27, 26],
    'designation': ['VP', 'CEO', 'CFO', 'VP', 'VP', 'CEO', 'CFO']
}

# Converting python dictionary into dataframe
df = pd.DataFrame(mydic)
df
```

```python
# 2. Converting dataframe into csv file
'''The Pandas to_csv() function is used to convert the DataFrame into CSV data.'''

# Syntax --> DataFrame.to_csv('csv_file_name', index=False)

# (If the value is set to True, index is included in the CSV data.
# Otherwise, the index value is not written in CSV output.)

df.to_csv('practical_1_csvfile', index=False)
df
```

```python
# 5. Reading Excel files

# syntax --> pandas.read_excel('/excel_file_location')
read_excelfile = pd.read_excel('FDS Datasets/gradedata.xlsx')
read_excelfile
```

```python
# changing column names of excel file
# syntax --> Dataframe.columns = ['name1', 'name2', 'name ....']

read_excelfile.columns = ['first_name', 'last_name', 'sex', 'age', 'exer', 'hrs', 'marks', 'add']
read_excelfile
```

```python
# 6. Writing dataframe into excel file

# create dataframe using list or any other datatype
names = ['Suraj', 'Aman', 'Alok', 'Vaishnavi']
grades= [75, 99, 90, 87]
bs_degree = [3, 1, 2, 4]
ms_degree = [1, 3, 4, 2]
degrees = zip(names, grades, bs_degree, ms_degree)
col = ['Names', 'Grades', 'BS', 'MS']
df = pd.DataFrame(data = degrees, index=['a', 'b', 'c', 'd'], columns=col)

# writing in excel file
'''
To write a single object to the excel file, we have to specify the target file name. 
If we want to write to multiple sheets, we need to create an ExcelWriter object with target filename 
and also need to specify the sheet in the file in which we have to write.
'''
# syntax --> pandas.ExcelWriter('file_name.xlsx')
writer = pd.ExcelWriter('practical_1_excelfile.xlsx', engine='xlsxwriter')

'''We can export the DataFrame to the excel file by using the to_excel() function.'''
# syntax --> DataFrame.to_excel(excel_writer, sheet_name='')
'''
excel_writer: A file path or existing ExcelWriter.
sheet_name: It refers to the name of the sheet that contains the DataFrame.
'''
df.to_excel(writer, sheet_name='Sheet1')

# finally save the file
writer.save()
```

```python
# 7. Adding dataframe into database

cars={'Brands':['Honda','Suzuki','Bugatti','Mazda'],
      'Price':[290000,300000,40000000,120000000]}
df=pd.DataFrame(data=cars,columns=('Brands','Price'))
con = sqlite3.connect('Test1.db')
c = con.cursor()
c.execute('Create Table casr_table(brands, price)')
con.commit()
```

```python
df.to_sql('CARS', con, if_exists='replace', index=False)
```
